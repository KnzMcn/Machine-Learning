**ANSWER:** The three strongest rules for Eclat are the minimum support, maximum combination, and minimum combination.

```python
#import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import files
import io

from pyECLAT import ECLAT
```

```python
#import the dataset
upload = files.upload()
dataset = pd.read_csv(io.BytesIO(upload['Market_Basket_Optimisation.csv']), header = None)
```

```python
dataset.shape
```

```python
# transfer the csv to an array
transactions = []
for i in range(7500):
  transactions.append([str(dataset.values[i, j]) for j in range(0, 20)])
```

```python
# convert the array into a dataframe
transactions = pd.DataFrame(transactions)

# show the dataset
transactions
```

# Train the model

```python
# instatiate the model
eclat = ECLAT(data = transactions, verbose = True)
```

```python
# set the hyperparameters to use

# minimum number of itemset size; meaning no individual products
min_n_products = 2 

# not sure about the support but the safest would be three, i guess
min_support = (3*7) / len(transactions)

# longest transaction in transactions
# 'len(x) for x in transactions' doesn't work so max_length will be 2, similar to the example
max_length = 2
```

```python
#fit the model
rule_indices, rule_supports = eclat.fit(min_support = min_support, 
                                        min_combination = 2, 
                                        max_combination = max_length)
```

```python
# show the results
results = list(rule_supports)
results[: 5]
```
